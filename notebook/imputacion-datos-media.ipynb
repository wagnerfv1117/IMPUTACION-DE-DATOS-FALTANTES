```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('C:/Users/WAGNER FERNÁNDEZ/OneDrive - POLICIA NACIONAL DE COLOMBIA/Documents/Ciencia de Datos/PROYECTOS/IMPUTACION-DE-DATOS FALTANTES/raw/offers.csv')
df.head(5)
```

```python
# Imputar valores faltantes en columnas numéricas con la media
df_numeric = df.select_dtypes(include='number')
df_numeric_imputed = df_numeric.apply(lambda col: col.fillna(col.mean()), axis=0)

# Imputar valores faltantes en columnas alfanuméricas con un valor específico (por ejemplo, 'Sin Información')
df_non_numeric = df.select_dtypes(exclude='number')
df_non_numeric_imputed = df_non_numeric.fillna('Sin Información')

# Concatenar las columnas imputadas
df_imputed = pd.concat([df_non_numeric_imputed, df_numeric_imputed], axis=1)

#Guardar el DataFrame imputado en un nuevo archivo CSV
df_imputed.to_csv('C:/Users/WAGNER FERNÁNDEZ/OneDrive - POLICIA NACIONAL DE COLOMBIA/Documents/Ciencia de Datos/PROYECTOS/IMPUTACION-DE-DATOS FALTANTES/output/offers_imputed_mean.csv', index=False)
```
